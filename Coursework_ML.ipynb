# 1. Описание данных

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#загрузка данных
full_df = pd.read_csv('telecom_users.csv')

full_df.head()
```

```python
#проверка наличия дубликатов по клинтам
full_df['customerID'].duplicated().sum()
```

```python
#убираем ненужные для задачи столбцы
full_df = full_df.drop(['Unnamed: 0', 'customerID'], axis = 1)
full_df.head()
```

```python
#проверим соотношение кол-ва отточных клиентов и не отточных
full_df['Churn'].value_counts()
```

Распреденение классов неравномерное. Не отточных пользователей в выборке больше чем в 2,5 раза больше, чем отточных.

```python
#визуализация распределения оттока клиентов по разным критериям - пол клиента, 
#состоит ли он в браке, имеет ли иждивенцев и является ли пенсионером
f, axes = plt.subplots(2, 2, figsize=(15,10))

sns.countplot(x='gender', hue='Churn', data=full_df[['Churn', 'gender']], ax=axes[0,0])
sns.countplot(x='Partner', hue='Churn', data=full_df[['Churn', 'Partner']], ax=axes[0,1])
sns.countplot(x='Dependents', hue='Churn', data=full_df[['Churn', 'Dependents']], ax=axes[1,0])
sns.countplot(x='SeniorCitizen', hue='Churn', data=full_df[['Churn', 'SeniorCitizen']], ax=axes[1,1])

plt.tight_layout()
```

Из графиков видно, что по полу разница среди клиентов практически одинаковая.
Пожилых людей среди клиентов заметно меньше, чем более молодых, так же и с наличием иждивенцев. 
Однако соотношение оттока и не оттока в группах с иждивенцами и нет - выглядит одинаковым.
Что касается групп состоящих в браке клиентов и свободных, кол-во одной и другой группы близко друг другу, но соотношение клиентов оттока и не оттока отличается.

```python
full_df.groupby('Churn')['tenure'].median()
```

По барчарту и цифрам видно, что отточные пользователи были клиентами в среднем гораздо меньше месяцев, чем не оточные.

# 2. Исследование зависимостей и формулирование гипотез

```python
#преобразование категориальных признаков в числовой формат, для удобства расчетов
full_df['gender'] = full_df['gender'].map( {'Female': 0, 'Male': 1} ).astype(int)

columns = ['Churn', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
for column in columns:
    full_df[column] = full_df[column].map( {'No': 0, 'Yes': 1} ).astype(int)
    
columns = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
for column in columns:
    full_df[column] = full_df[column].map({'No': 0, 'Yes': 1, 'No internet service': 2}).astype(int)
    
full_df['MultipleLines'] = full_df['MultipleLines'].map({'No': 0, 'Yes': 1, 'No phone service': 2}).astype(int)
full_df['InternetService'] = full_df['InternetService'].map({'No': 0, 'DSL': 1, 'Fiber optic': 2}).astype(int)
full_df['Contract'] = full_df['Contract'].map({'Month-to-month': 0, 'One year': 1, 'Two year': 2}).astype(int)
full_df['PaymentMethod'] = full_df['PaymentMethod'].map(
    {'Credit card (automatic)': 0, 'Bank transfer (automatic)': 1,
     'Electronic check': 2, 'Mailed check': 3}).astype(int)
```

### Выбор признаков для обучения моделей прогнозирования.

Убираем все интернет-услуги (OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport, StreamingTV, StreamingMovies), так как по тепловой карте видна сильная корреляция с типом интернет провайдера - InternetService. Значимая связь между всеми интернет услугами и в целом подключен ли интернет у клиента или нет. Данные взаимосвязи могут исказить показания моделей.

Так же с признаком InternetService сильно коррелирует размер месячной оплаты (MonthlyCharges), его тоже исключим. Очевидно, что если у клиента есть интернет, то оплата его будет выше, учитывая и дополнительные услуги.

Есть связь между PhoneService и MultipleLines, уберем второе. Очевидно, что если подключена телефонная линия, то их может быть несколько у клиента.

Так же связаны между собой тип контракта и кол-во месяцев, которые человек является клиентом. Однако сам по себе признак tenure достаточно различается по группам оттока, поэтому оставим эти два признака в данных для моделей.

Используя анализ групп по социальным характеристикам клиентов, оставим признак Partner и SeniorCitizen. В них распределение внутри групп пользователей оттока наиболее интересно.

```python
#создание переменных с данными
data = full_df[['Partner', 'tenure', 'PhoneService', 'InternetService', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'TotalCharges']]
target = full_df.Churn

data.head()
```

# 3. Построение моделей для прогнозирования оттока

Задача прогнозирования уйдет ли клиент в отток или нет - это задача бинарной классификации.
Для этого больше всего подходят модели дающие вероятностные предсказания. Это позволяет в какой-то степени избежать влияния несбалансированности классов на результаты алгоритмов.
Для прогнозирования выбраны несколько алгоритмов:
- Наивный классификатор Байеса
- Логистическая регрессия
- Random Forest классификатор
- Градиентный бустинг
- Стекинг на основе базовых алгоритмов и градиентного бустинга в качестве мате-модели

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import svm
from sklearn.svm import SVC
```

```python
#разделение данных на тренировочную и тестовую выборку
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.25, random_state=10)
```

```python
#обучение модели и получение предсказаний классификатором Байеса
gnb = GaussianNB()
gnb_model = gnb.fit(x_train, y_train)

gnb_probas = gnb_model.predict_proba(x_test)
gnb_pred = gnb_model.predict(x_test)

gnb_pred.shape
```

```python
#обучение модели и получение предсказаний логистической регрессии
logreg = LogisticRegression(max_iter = 1000, random_state=10)
logreg_model = logreg.fit(x_train, y_train)

logreg_probas = logreg_model.predict_proba(x_test)
logreg_pred = logreg_model.predict(x_test)

logreg_pred.shape
```

```python
#построение модели Random Forest и поиск наилучших параметров с помощью RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
rfc = RandomForestClassifier(random_state = 10)

params = {'n_estimators': range(10, 100, 5), 
          'max_depth': range(1, 11), 
          'min_samples_leaf': range(1, 50), 
          'min_samples_split': range(10, 100, 5)}

clf = RandomizedSearchCV(rfc, params, cv = 5, scoring='roc_auc').fit(x_train, y_train)

clf.best_params_
```

```python
#обучение лучшего алгоритма и получение предсказаний на тестовой выборке
rfc_model = clf.best_estimator_.fit(x_train, y_train)

rfc_pred = rfc_model.predict(x_test)
rfc_probas = rfc_model.predict_proba(x_test)

rfc_pred.shape
```

```python
#построение модели градиентного бустинга и поиск наилучших параметров с помощью RandomizedSearchCV
gbc = GradientBoostingClassifier(random_state = 10)

params = {'n_estimators': [1] + list(range(10, 100, 10)), 
          'learning_rate': np.linspace(0.01, 0.99, 25)}

clf = RandomizedSearchCV(gbc, params, scoring='roc_auc', cv = 5).fit(x_train, y_train)

clf.best_params_
```

```python
#обучение алгоритма с лучшими параметрами и получение предсказания
gbc_model = clf.best_estimator_.fit(x_train, y_train)

gbc_pred = gbc_model.predict(x_test)
gbc_probas = gbc_model.predict_proba(x_test)

gbc_pred.shape
```

Для построения ансамбля стекинга в качестве базовых алгоритмов были выбраны логистическая регрессия, дерево решений, классификатор Байеса и метод опорных векторов.
Мета-алгоритм - градиентный бустинг.

```python
#разбиение тренировочных данных дополнительно на тренировочную и валидационную выборки
train, valid, train_true, valid_true = train_test_split(x_train, 
                                                        y_train,
                                                        train_size=0.5,
                                                        random_state=10)
```

```python
#создание базовых алгоритмов и обучение подтренировочной выборке
lr = LogisticRegression(max_iter = 1000, random_state=10)
lr_model = lr.fit(train, train_true)

dtc = DecisionTreeClassifier(max_leaf_nodes=4, random_state=10)
dtc_model = lr.fit(train, train_true)

gnb = GaussianNB()
gnb_model = gnb.fit(train, train_true)

svc = svm.SVC(class_weight={1: 3})
svc_model = svc.fit(train, train_true)
```

```python
#получение матрицы признаков для мета-алгоритма из предсказаний базовых алгориимов на валидационной подвыборке
#вывод метрик Roc-Auc для оценки внутренних результатов базовых алгоритмов
models = [lr_model, dtc_model, gnb_model, svc_model]
meta_mtrx = np.empty((valid.shape[0], len(models)))

for n, model in enumerate(models):
    meta_mtrx[:, n] = model.predict(valid)
    predicted = model.predict(x_test)
    print(f'{n} auc: {roc_auc_score(y_test, predicted)}')

#обучение мета-алгоритма на матрице признаков    
meta = GradientBoostingClassifier(n_estimators=40)
meta_model = meta.fit(meta_mtrx, valid_true)
```

```python
#построение мета-матрицы признаков на предсказаниях бозовых моделей по тестовой выборке
meta_mtrx_test = np.empty((x_test.shape[0], len(models))) 

for n, model in enumerate(models):
    meta_mtrx_test[:, n] = model.predict(x_test)

#получение предсказаний и общей метрики Roc-Auc стекинга    
meta_predict = meta.predict(meta_mtrx_test)
meta_proba = meta.predict_proba(meta_mtrx_test)
print(f'Stacking AUC: {roc_auc_score(y_test, meta_predict)}')
```

# 4. Сравнение качества моделей 

Для сравнения качества моделей возьмем метрику ROC-AUC, так как она хорошо отражает насколько верно модель прогнозирует оба класса.
Для сравнения так же посмотрим на полноту. В данном случае важно не упускать действительно отточных пользователей, то есть минимизировать ложно-отрицательные прогнозы. Чтобы успеть предпринять меры по удержанию таких клиентов.

```python
from sklearn.metrics import recall_score, roc_auc_score, roc_curve
```

```python
#расчет метрики ROC-AUC для всех моделей
gnb_roc_auc = roc_auc_score(y_test, gnb_pred)
logreg_roc_auc = roc_auc_score(y_test, logreg_pred)
rfc_roc_auc = roc_auc_score(y_test, rfc_pred)
gbc_roc_auc = roc_auc_score(y_test, gbc_pred)
stack_roc_auc = roc_auc_score(y_test, meta_predict)

print('Метрика ROC-AUC')
print(f'GaussianNB: {gnb_roc_auc},\nLogisticRegression: {logreg_roc_auc}, \nRandomForestClassifier: {rfc_roc_auc}')
print(f'GradientBoostingClassifier: {gbc_roc_auc}, \nStacking model: {stack_roc_auc}')
```

```python
models = ['GNB', 'LOG_REG', 'RFC', 'GBC', 'STACK']
roc_auc_scores = [gnb_roc_auc, logreg_roc_auc, rfc_roc_auc, gbc_roc_auc, stack_roc_auc]

sns.barplot(x=models, y=roc_auc_scores, palette="Blues_d")
```

Лучший показатель по метрике ROC-AUC у классификатора Байеса с дефолтными параметрами. Чуть ниже показатель у ансамбля стекинга. Кстати в рамках базовых алгоритмов внутри модель Байеса так же показывает лучшее значение.
Хуже всех с прогнозом справляется случайный лес. Но разница в показателях моделей относительно друг друга невысокая.

```python
#построение графиков roc_auc для каждой модели
f = plt.figure(figsize=(10,7))

fpr, tpr, _ = roc_curve(y_test, gnb_probas[:, 1])
plt.plot(fpr, tpr, label="auc_gnb=" + str(gnb_roc_auc))

fpr, tpr, _ = roc_curve(y_test, logreg_probas[:, 1])
plt.plot(fpr, tpr, label="auc_logreg=" + str(logreg_roc_auc))

fpr, tpr, _ = roc_curve(y_test, rfc_probas[:, 1])
plt.plot(fpr, tpr, label="auc_rfc=" + str(rfc_roc_auc))

fpr, tpr, _ = roc_curve(y_test, gbc_probas[:, 1])
plt.plot(fpr, tpr, label="auc_gbc=" + str(gbc_roc_auc))

fpr, tpr, _ = roc_curve(y_test, meta_proba[:, 1])
plt.plot(fpr, tpr, label="auc_stack=" + str(stack_roc_auc))

plt.legend(loc=4)

plt.show()
```
