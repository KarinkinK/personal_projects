```python
#импорт нужных библиотек
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from pandas.plotting import register_matplotlib_converters
%matplotlib inline

register_matplotlib_converters()

#чтение датасетов из файлов
courses = pd.read_csv('courses.csv')
students = pd.read_csv('students.csv')
course_contents = pd.read_csv('course_contents.csv')
progresses = pd.read_csv('progresses.csv')
progress_phases = pd.read_csv('progress_phases.csv')

#подготовка таблиц для корректного объединения
del courses['Unnamed: 0']
courses.columns = ['course_id', 'course_title', 'course_field']
progresses.columns = ['progress_id', 'student_id', 'course_id']
```

```python
#объединение датасетов
full_df = pd.merge(progress_phases, progresses, how = 'left', left_on='progress_id', right_on='progress_id') 
full_df = pd.merge(full_df, courses, how = 'left', left_on='course_id', right_on='course_id')
full_df = pd.merge(full_df, course_contents, how = 'left', 
                   left_on=['lesson_number', 'module_number', 'course_id'], 
                   right_on=['lesson_number', 'module_number', 'course_id'])

full_df.info() #информация по итоговой таблице
full_df.head()
```

```python
#проверка итогового датасета 
full_df[full_df.student_id == '768c2987a744c51ce64a5993a2a94eaf']
```

```python
#создание таблицы с данными о возрасте и городе студентов
del students['id_']
students.columns = ['student_id', 'city', 'birthday']
stud_course = pd.merge(students, progresses, how = 'left', left_on='student_id', right_on='student_id')
stud_course.dropna(subset = ['course_id'], inplace = True)
stud_course = pd.merge(stud_course, courses, how = 'left', left_on='course_id', right_on='course_id')
stud_course.head(15)
```

```python
#1.1 общее количество курсов в датасете
sum_courses = len(full_df['course_id'].unique())
print('Общее количество курсов:', sum_courses)
print('------------------------')

#количество модулей на каждом курсе
sum_module = full_df.groupby('course_title')['module_title'].unique().agg(len)
print('Количество модулей на каждом курсе:\n', sum_module)
```

```python
#1.1 количество уроков в каждом модуле на каждом курсе
sum_lesson = full_df.groupby(['course_title', 'module_title'])
sum_lesson = sum_lesson['lesson_title'].nunique().reset_index(name = 'sum_lesson')
print('Количество уроков в каждом модуле:')
sum_lesson.head(20)
```

```python
#1.1 медианное количество уроков в модуле на каждом курсе
med_lesson = sum_lesson.groupby(['course_title'])['sum_lesson'].median()
print('Медианное количество уроков в модуле на каждом курсе:')
med_lesson
```

```python
#1.1 количество учеников на каждом курсе
sum_students = full_df.groupby(['course_field', 'course_title'])['student_id'].nunique().reset_index(name = 'sum_students')
print('Количество студентов на каждом курсе:\n', sum_students)
```

```python
#подготовка таблицы для расчетов данных по студентам

stud_course['birthday'] = pd.to_datetime(stud_course['birthday'], errors = 'coerce', format='%Y-%m-%d')
stud_course.dropna(subset = ['birthday'], inplace = True)
stud_course['age'] = (pd.to_datetime('now') - stud_course['birthday']).astype('<m8[Y]')
stud_course['age'] = stud_course['age'].astype(int)

#так как многие посчитанные значения возрастов были неккоректные,
#был взят диапазон наиболее объективного возраста студентов от 12 до 65 лет
stud_course1 = stud_course[(12 <= stud_course['age']) & (65 >= stud_course['age'])]
stud_course1.head(10)
```

```python
#1.1 минимальный, максимальный, средний, медианный возраст студентов
min_age = stud_course1['age'].min()
print('Минимальный возраст студентов:', min_age)
print('------------------------------')
max_age = stud_course1['age'].max()
print('Максимальный возраст студентов:', max_age)
print('------------------------------')
mean_age = round(stud_course1['age'].mean())
print('Средний возраст студентов:', mean_age)
print('------------------------------')
med_age = round(stud_course1['age'].median())
print('Медианный возраст студентов:', med_age)
print('------------------------------')

#1.1 минимальный, максимальный, средний, медианный возраст студентов на каждом курсe
min_age_course = stud_course1.groupby(['course_title'])['age'].min()
print('Минимальный возраст на курсе:\n', min_age_course)
print('------------------------------')
max_age_course = stud_course1.groupby(['course_title'])['age'].max()
print('Максимальный возраст на курсе:\n', max_age_course)
print('------------------------------')
mean_age_course = round(stud_course1.groupby(['course_title'])['age'].mean())
print('Средний возраст на курсе:\n', mean_age_course)
print('------------------------------')
med_age_course = round(stud_course1.groupby(['course_title'])['age'].median())
print('Медианный возраст на курсе:\n', med_age_course)
print('------------------------------')
```

```python
#1.2 Постройте bar-chart, отражающий количество студентов на каждом курсе.
fig = plt.figure()

x = sum_students['course_title']
y = sum_students['sum_students']

plt.bar(x, y)
plt.title('Количество студентов на каждом курсе')
plt.xticks(rotation = 90)

plt.show()
```

```python
#1.3 Горизонтальный bar-chart, отражающий количество студентов на каждом курсе
fig = plt.figure(figsize = (8, 6))

sort_sum_stud = sum_students.sort_values(by='sum_students', ascending = False)
med = int(sort_sum_stud['sum_students'].median())

ax = fig.add_axes([0.05, 0.05, 0.9, 0.9])
ax.set_title('Количество студентов на каждом курсе')

line = ax.vlines(med, 0, 14.5, color = 'r', label = 'median = 782')

for name in sort_sum_stud['course_field'].unique():
    x = sort_sum_stud[sort_sum_stud.course_field == name]['course_title']
    y = sort_sum_stud[sort_sum_stud.course_field == name]['sum_students']
    ax.barh(x, y, alpha = 0.1, label = name)
    
ax.legend()

ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False) 
ax.spines['bottom'].set_visible(False) 
ax.spines['left'].set_visible(False) 

plt.show()
```

```python
#2.1 Расчет прироста студентов на каждом курсе в каждом месяце за всю историю

#делаем датасет с данными по времени начала и конца всех домашних работ
teacher_load = full_df[['status', 'start_date', 'finish_date', 'student_id',
                        'course_title', 'module_title', 'lesson_title', 'is_homework']].copy()

     
teacher_load['start_date'] = pd.to_datetime(teacher_load['start_date'], errors = 'coerce', format='%Y-%m-%d')
teacher_load['finish_date'] = pd.to_datetime(teacher_load['finish_date'], errors = 'coerce', format='%Y-%m-%d')

teacher_load = teacher_load[teacher_load['is_homework'] == True]

teacher_load.info()
teacher_load
```

```python
#создание таблицы с датой первой домашней работы для каждого студента на каждом курсе
growt_stud = teacher_load.groupby(['course_title', 'student_id'])['start_date'].apply(np.min).reset_index()
```

```python
#создание столбца месяц-дата для удобного подсчета прироста кол-ва студентов на каждом курсе

growt_stud['start_date'] = pd.to_datetime(growt_stud['start_date'].dt.strftime('%m/%Y'))

#Расчет прироста студентов на каждом курсе в каждом месяце за всю историю
growt_month = growt_stud.groupby('course_title')['start_date'].value_counts().reset_index(name = 'growth')
growt_month = pd.pivot_table(growt_month, index=growt_month['start_date'], values = 'growth', 
                             columns = growt_month['course_title'], fill_value=0)


growt_month
```

```python
#2.2 line-graph'ы прироста студентов на каждом отдельном курсе

titles = list(growt_month.columns)

fig, subplots = plt.subplots(nrows=5, ncols=3, figsize = (12, 15))

i=-1
for ax in fig.axes:
    i += 1
    ax.set_title(titles[i])
    x = growt_month.index
    y = growt_month[titles[i]]
    ax.plot(x, y)
    ax.set_xlabel('Months')
    ax.set_ylabel('Growth')
    ax.tick_params(axis='x', labelrotation = 45)
 

plt.tight_layout(h_pad = 2)

plt.show()    
```

```python
#список из дат месяц-год за все время наблюдения
list_months = pd.date_range('2016-03-01', '2019-07-01', freq='MS').strftime('%Y-%m-%d')

#пустой датафрейм из одной строки, чтобы к нему добавлять строки с посчитанным кол-вом прогрессов
progress_month1 = pd.DataFrame(0, index = ['0'], columns=growt_month.columns)

#цикл подсчета прогресса для каждого месяца с группировкой по курсам, добавление строк к пустому датафрейму
for date in list_months:
    progress_temp = progress_hw1[(progress_hw1['finish_date'].dt.strftime('%Y-%m-%d') >= date) 
                             & (progress_hw1['start_date'].dt.strftime('%Y-%m-%d') <= date)]

    progress_temp = progress_temp.groupby('course_title').size().reset_index(name = date)
    progress_temp = pd.pivot_table(progress_temp, values = date, 
                                   columns = progress_temp['course_title'], fill_value=0)
    
    progress_month1 = progress_month1.append(progress_temp, sort=False)

#удаление первой строчки и замена всех nan на 0    
progress_month1 = progress_month1.drop(['0'])
progress_month1 = progress_month1.fillna(0).astype(int)


progress_month1 #таблица прогрессов выполнения дз со статусом done
```

```python
#4.1 Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе.

#выбираем с общего датафрейма нужные столбцы, фильтруем по дз и приводим дату к удобному виду
conversion_df = full_df[['status', 'start_date', 'finish_date', 'student_id',
                        'course_title', 'module_title', 'module_number', 'is_homework']].copy()

conversion_df = conversion_df[conversion_df['is_homework'] == True]

conversion_df['start_date'] = pd.to_datetime(conversion_df['start_date']).dt.date
conversion_df['finish_date'] = pd.to_datetime(conversion_df['finish_date']).dt.date

conversion_df.info()
conversion_df.head()
```

```python
#создаем таблицу с первыми дз у каждого студента на каждом модуле и курсе

count_start_module = conversion_df.groupby(['course_title', 'module_title', 'module_number', 'student_id'])['start_date'].min().reset_index()
count_start_module.head(10)
```

```python
#считаем колонку с количеством приступивших к каждому модулю студентов

count_start_module = count_start_module.groupby(['course_title', 'module_title', 'module_number'])['student_id'].count().reset_index(name = 'count_start')
count_start_module = count_start_module.groupby('course_title').apply(lambda x:x.sort_values(by='module_number')).reset_index(drop=True)
count_start_module.head(10)
```

```python
#создаем таблицу с датами, когда было принято последнее в каждом модуле дз

count_finish_module = conversion_df[conversion_df['status'] == 'done']

count_finish_module = count_finish_module.groupby(['course_title', 'module_title', 'module_number', 'student_id'])['finish_date'].max().reset_index()
count_finish_module.head(10)
```

```python
#считаем колонку с количеством завершивших каждый модуль студентов

count_finish_module = count_finish_module.groupby(['course_title', 'module_title', 'module_number'])['student_id'].count().reset_index(name = 'count_finish')
count_finish_module = count_finish_module.groupby('course_title').apply(lambda x:x.sort_values(by='module_number')).reset_index(drop=True)

count_finish_module.head(10)
```

```python
#объединяем две предыдущие таблицы в одну, в ней ыидим сколько студентов приступило к модулю и сколько его завершило

conversion_module = pd.merge(count_start_module, count_finish_module, how = 'left',
                             left_on=['course_title', 'module_title', 'module_number'], right_on=['course_title','module_title', 'module_number'])

conversion_module.head(10)
```

```python
#считаем конверсию для каждого модуля на каждом курсе

title_list = conversion_module['course_title'].unique().tolist()

conversion_module['conversion'] = 0

for title in title_list:   #цикл, которые выбирает строчки для каждого отдельного курса
    a = conversion_module['count_start'][conversion_module['course_title'] == title][1:] #значения кол-ва начавших модуль, не считая самый первый
    b = conversion_module['count_finish'][conversion_module['course_title'] == title][:-1] #значения кол-ва завершивших модуль, без последнего
    b.index = b.index + 1 #увеличение индекса, чтобы считалось отношение текущего к предыдущему
    conversion_module.loc[conversion_module.course_title == title, 'conversion'] = round(a/b, 2) #переопределение 0 на значение посчитанной конверсии
    
conversion_module.head(25)
```

```python
#фильтруем таблицу, чтобы убрать все первые модули в курсах, т.к. конверсия для них не считалась
conversion_module = conversion_module[conversion_module['module_number'] != 1]

#проверяем еще раз наличие значений Nan
conversion_module[conversion_module['conversion'].isna()]
```
